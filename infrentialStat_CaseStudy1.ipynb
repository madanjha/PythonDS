**Statical Analysis**
1) preparing a sample
2) creating hypothesis(Null and alternate hypothisis)
3) Applying the appropiate test based on requirement
4) compare the p_value going to decide the accept or reject.




```python
#Q it was obeserve that the avg purchage made by man of the age 18-25 was10000 . is it still the same?
## Extract the data for man age of group
new_data=data[(data["Age"]==1) & (data["Gender"]==1)]
new_data
```

```python
sample=new_data.sample(3600,random_state=5)
sample
```

```python
#null_hypth=mean will be equal 10000
#alt_hypth=means will not equal to 10000

new_data['Purchase'].mean()
```

```python
from scipy.stats import ttest_1samp
alph=0.05
a_mean=10000
t_stat, p_value=ttest_1samp(sample['Purchase'],a_mean)
print(t_stat)
print(p_value)
value=1.7977812377777705e-10
print(f"{value:.25f}")
if p_value<alph:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
