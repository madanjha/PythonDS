#### **To solve above mentioned problem we have to use tools like `iloc` and `loc`**.

#### **iloc : index based selection of the data**

```python
market.iloc[10]
```

```python
market.iloc[[10, 76, 98, 345, 889, 1043]]
```

```python
market.iloc[500:1001, [5, 3, 7]]
```

```python
#### loc: label(name) based selection of the data
```

```python
market.loc[500:1000, ['Discount','Cust_id', 'Profit']]
```

```python
market.loc[[10, 65, 98, 67, 990]]
```

### **Selection of the data using conditions**

```python
market.head(15)
```

```python
market.iloc[[0, 5, 9 13, 14, .....]]
```

```python
market['Profit'] < 0  #go and check which row has negative profit
```

```python
market[market['Profit'] < 0]
```

```python
market['Order_Quantity'] > 10
```

```python
market[market['Order_Quantity'] > 10]
```

```python
(market['Profit'] > 10) & (market['Profit'] < 50)
```

```python
market[(market['Profit'] > 10) & (market['Profit'] < 50)]
```

```python
(market['Profit'] < 0) & (market['Sales'] > 1000)
```

```python
market[(market['Profit'] < 0) & (market['Sales'] > 1000)]
```
