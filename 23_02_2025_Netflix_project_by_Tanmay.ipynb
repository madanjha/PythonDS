# **Netflix Recomendation Project**


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix_dataset=pd.read_csv("/content/drive/MyDrive/Netflix dataset/combined_data_1.txt.zip",header=None,usecols=[0,1],names=["Cust_ID","Rating"])
netflix_dataset
```

```python
# How many Total number of movies
```

```python
movie_count=netflix_dataset.isnull().sum()["Rating"]
```

```python
movie_count
```

```python
movie_id=None
movie_list=[]
for cust_id in netflix_dataset["Cust_ID"]:
  if ':' in cust_id:
    movie_id=int(cust_id.replace(":",""))
  movie_list.append(movie_id)
```

```python
movie_list
```

```python
netflix_dataset["Movie_ID"]=movie_list
```

```python
netflix_dataset
```

```python
netflix_dataset.dropna(inplace=True)
```

```python
netflix_dataset
```

```python
# Pre filtering
```

```python
# movie filter
# remove the movies with less number of ratings
data_movie_summary=netflix_dataset.groupby("Movie_ID")["Rating"].count()
data_movie_summary
```

```python
#setting up benchmark
movie_benchmark=round(data_movie_summary.quantile(0.60))
movie_benchmark
```

```python
Rejected_movies=data_movie_summary[data_movie_summary<movie_benchmark].index
Rejected_movies
```

```python
# Customer Filter
# removing the customers who have rated less number of movies
```

```python
data_cust_summary=netflix_dataset.groupby("Cust_ID")["Rating"].count()
```

```python
data_cust_summary
```

```python
#setting up benchmark
cust_benchmark=round(data_cust_summary.quantile(0.60))
cust_benchmark
```

```python
Rejected_cust=data_cust_summary[data_cust_summary<cust_benchmark].index
Rejected_cust
```

```python
#Removing the rejected movies and rejected customers from the data


# ~ is exclusion or negation operator which will exclude the resulted data from the actual data
```

```python
netflix_dataset=netflix_dataset[~netflix_dataset["Movie_ID"].isin(Rejected_movies)]
```

```python
netflix_dataset=netflix_dataset[~netflix_dataset["Cust_ID"].isin(Rejected_cust)]
```

```python
netflix_dataset
```

```python
netflix_dataset.info()
```

```python
netflix_dataset["Cust_ID"]=netflix_dataset["Cust_ID"].astype(int)
```

```python
netflix_dataset
```

```python
from surprise import SVD,Dataset,Reader
from surprise.model_selection import cross_validate
```

```python
read=Reader()
```

```python
data=Dataset.load_from_df(netflix_dataset[["Cust_ID","Movie_ID","Rating"]][:1000000],read)  # performing on 10 lakhs rows only for quick analyis
# load_from_df - It will be converting the data into SVD readable object format
```

```python
# model Building
model=SVD()
```

```python
cross_validate(model,data,measures=["RMSE"],cv=3)
```
