```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_excel('/content/BankSetup.xlsx')
df.info()
df
for col in df.columns:
    if df[col].dtype == 'int64':
        df[col] = df[col].astype('object')
df.info()
```

Subtask:
Write each JSON string to a separate file named after the branchNumber code.

Reasoning: Iterate through the list of JSON strings and write each to a file named after the branchNumber code.Also while converting the field value to json all value should be take as text and structure like below
 {
"branchNumber": "AB213"
"branchData":{
   "branchCode":{
   value": "403"
   },

```python
import json
import os

# Create a directory to store the JSON files if it doesn't exist
if not os.path.exists('bank_data'):
    os.makedirs('bank_data')

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    # Convert the row to a dictionary
    row_dict = row.to_dict()

    # Get the branchNumber for the filename
    branch_number = str(row_dict['branchNumber'])

    # Create the desired JSON structure
    json_output = {
        "branchNumber": branch_number,
        "branchData": {
            "branchCode": {"value": str(row_dict.get('branchData.branchCode.value', ''))},
            "BANK": {"value": str(row_dict.get('BANK.value', ''))},
            "IFSC": {"value": str(row_dict.get('IFSC.value', ''))},
            "BRANCH": {"value": str(row_dict.get('BRANCH.value', ''))},
            "ADDRESS": {"value": str(row_dict.get('ADDRESS.value', ''))},
            "CITY1": {"value": str(row_dict.get('CITY1.value', ''))},
            "CITY2": {"value": str(row_dict.get('CITY2.value', ''))},
            "STATE": {"value": str(row_dict.get('STATE.value', ''))},
            "STD CODE": {"value": str(row_dict.get('STD CODE.value', ''))},
            "PHONE": {"value": str(row_dict.get('PHONE.value', ''))},
        }
    }

    # Convert the dictionary to a JSON string
    json_string = json.dumps(json_output, indent=4)

    # Define the filename
    filename = f'bank_data/{branch_number}.json'

    # Write the JSON string to the file
    with open(filename, 'w') as f:
        f.write(json_string)

print("JSON files created successfully with the updated structure!")
```
