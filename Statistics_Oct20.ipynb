```python
age=[18,25,30,45,33,-190,67,78,89,93,83,250,76,99,271]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(age)
plt.show()
```

```python
# IQR=Q3-Q1
```

```python
Q1=np.percentile(age,25)
```

```python
Q3=np.percentile(age,75)
```

```python
IQR=Q3-Q1
```

```python
IQR
```

```python
# Upper_fence and Lower_fence
```

```python
UL=Q3+1.5*IQR
LL=Q1-1.5*IQR
```

```python
UL
```

```python
LL
```

```python
for i in age:
  if i>UL or i<LL:
    print('Outlier values in data is:',i,'And it has been deleted')
    age.remove(i)
print(age)
```

**Normal Distribution**

Bell shaped curve and values will be equally distributed in both sides.

Mean and median will be there in the centre of the data

```python
data=[15,16,16,17,17,18,18,19]
```

```python
plt.hist(data,bins=5,color='red',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
data=[10,11,11,12,12,12,12,12]
plt.hist(data,bins=3,color='red',edgecolor='black')
plt.title('Left skewed graph that will give -ve skewness values')
plt.show()
```

```python
from scipy.stats import skew
ans=skew(data)
ans
```

```python
data=[10,10,10,10,10,11,11,12]
plt.hist(data,bins=3,color='red',edgecolor='black')
plt.title('Right skewed graph that will give +ve skewness values')
plt.show()
```

```python
ans=skew(data)
ans
```
