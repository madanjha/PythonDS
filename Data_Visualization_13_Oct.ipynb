## **Types of Plots**

### **1. Scatter Plot (Bivariate Plot)**

- Represents a relationship between two numerical columns in a dataframe.

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
x = tips['total_bill']
y = tips['tip']

plt.scatter(x, y, color = 'orange', marker = '*', s = 100, edgecolors = 'black')   #create the plot
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total bill vs Tip')
plt.show()  #display the plot
```

### **2. Histogram (Univariate Plot)**

- Represents the frequency of data in some range. We use a numerical column to make a histogram.

```python
tips.head()
```

```python
x = tips['total_bill']

plt.hist(x)
plt.xlabel('Total Bill')
plt.ylabel('Count of Customers')
plt.show()
```

### **3. Pie Chart(Univariate Plot)**

```python
tips.head()
```

```python
tips.shape
```

```python
tips['smoker'].unique()
```

```python
tips['day'].value_counts()
```

```python
plt.pie(tips['day'].value_counts(), labels = ['Sat', 'Sun', 'Thur', 'Friday'], autopct = '%.2f%%',
        explode = [0, 0, 0, 0.4])
plt.show()
```

### **4. Bar Chart(Univariate Plot)**

```python
tips.head()
```

```python
tips['day'].unique()
```

```python
tips['day'].value_counts()
```

```python
plt.bar(['Sat', 'Sun', 'Thur', 'Fri'], height = tips['day'].value_counts(), width = 0.5)
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.show()
```

### **5. Figures and Subplots**

```python
plt.figure(figsize = (10, 5))

plt.subplot(1,2, 1)
plt.bar(['Sat', 'Sun', 'Thur', 'Fri'], height = tips['day'].value_counts(), width = 0.5)

plt.subplot(1,2, 2)
plt.pie(tips['day'].value_counts(), labels = ['Sat', 'Sun', 'Thur', 'Friday'], autopct = '%.2f%%', explode = [0, 0, 0, 0.4])

plt.show()
```

## **Data Visualization with Seaborn**

```python
import seaborn as sns
```

### **1. Scatter Plot**

```python
tips.head()
```

```python
sns.set()  #use the default styling of all further plots to be of seaborn style
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter', hue = 'smoker', col = 'sex', palette = ['green', 'cyan'])
plt.show()
```

### **2. Line Plot**

```python
flights = pd.read_csv('flights.csv')
```

```python
flights.head()
```

```python
sns.relplot(x = 'year', y = 'passengers', data = flights, kind = 'line')
plt.show()
```

### **3. Countplot (Similar to a bar chart)**

```python
tips.head()
```

```python
sns.countplot(x = 'day', data = tips, hue = 'sex')
plt.show()
```

### **4. Boxplot**

- Used to represent a relationship between one categorical and one numerical.

```python
tips.head()
```

```python
sns.boxplot(y = 'total_bill', x = 'sex', data = tips)
plt.show()
```

### **5. Regplot/Regression Plot**

```python
tips.head()
```

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

### **6. Distribution Plots**

#### **6.1 Kdeplot**

```python
tips.head()
```

```python
sns.kdeplot(x = 'total_bill', data = tips)
plt.show()
```

#### **6.2 Displot**

```python
sns.displot(x = 'total_bill', data = tips)
plt.show()
```
