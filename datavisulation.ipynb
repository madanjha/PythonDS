#**Types of Plots**

### 1. Scattere plot(Bivriate)

```python
tips=pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
x=tips['total_bill']
y=tips['tip']
plt.scatter(x,y, color='magenta',marker='*',s=100,edgecolor='black')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('Total bill vs Tip')
plt.show()
#
```

### **2. Histogram(Univariate Plot)**
- represent the freq of data in some range. we uses numirical column in range

```python
x = tips['total_bill']

plt.hist(x)
plt.xlabel('Total Bill')
plt.ylabel('count the cutomer')
plt.title('Histogram of Total Bill')
plt.show()
```

### **3. Pie chart (univariate Plot)**

```python
tips.head()
```

```python
tips.shape
```

```python
tips['sex'].unique()
```

```python
plt.pie(tips['day'].value_counts(),labels=['sat','sun','friday','thur'],autopct='%.2f%%',
          explode=[0,0,0,0.2])
plt.show()
#
```

### **4.Bar chart**

```python
tips.head()
```

```python
tips['day'].value_counts()
```

```python
plt.bar(['sat','sun','friday','thur'],height=tips['day'].value_counts())
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.show()
```

### **5. figure and subplots**

```python
plt.figure(figsize = (10, 5))

plt.subplot(1,2, 1)
plt.bar(['Sat', 'Sun', 'Thur', 'Fri'], height = tips['day'].value_counts(), width = 0.2)

plt.subplot(1,2, 2)
plt.pie(tips['day'].value_counts(), labels = ['Sat', 'Sun', 'Thur', 'Friday'], autopct = '%.2f%%', explode = [0, 0, 0, 0.4])

plt.show()
```

Data Visualization with Seaborn

```python
import seaborn as sns
```

**1. Scatter Plot**

```python
tips.head()
```

```python
sns.set()
sns.relplot(x='total_bill',y='tip',data=tips,kind='scatter',hue='sex',col='sex')
plt.show()
#
```

#**Line Plot**

```python
flights=pd.read_csv('flights.csv')
```

```python
flights.head()
```

```python
sns.relplot(x='year', y='passengers', data=flights, kind='line')
plt.show()
```

**3. Countplot (Similar to a bar chart**)

```python
tips.head()
```

```python
sns.countplot(x = 'day', data = tips, hue = 'sex')
plt.show()
```

**4. Boxplot**
Used to represent a relationship between one categorical and one numerical.

```python
tips.head()
```

```python
sns.boxplot(y = 'total_bill', x = 'sex', data = tips)
plt.show()
```

5. Regplot/Regression Plot

```python
tips.head()
```

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
plt.show()
```

6. Distribution Plots

6.1 Kdeplot
